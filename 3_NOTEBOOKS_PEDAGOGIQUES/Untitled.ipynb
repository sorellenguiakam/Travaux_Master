## SLIDE 3 : QU'EST-CE QUE LE DIABÈTE DE TYPE 1 ?
**[Durée : 2 minutes]**

« Le Diabète de Type 1 est une **maladie auto-immune chronique**. Cela signifie que le système immunitaire du patient — normalement protecteur — s'attaque par erreur à ses propres cellules. »

« Plus précisément, les cellules **bêta du pancréas** — responsables de produire l'insuline — sont progressivement détruites. L'insuline est l'hormone essentielle qui permet aux cellules de capturer le sucre dans le sang. »

« Quand ces cellules disparaissent, le patient ne produit plus d'insuline. Sans insuline, le glucose s'accumule dans le sang : c'est l'**hyperglycémie**. Le patient doit alors recevoir de l'insuline externe, **à vie**. »

*[Montrer les colonnes avec les chiffres]*

« Épidémiologiquement — et c'est important de comprendre l'ampleur — nous parlons de :

- **8,4 millions** de personnes diagnostiquées dans le monde
- **1 million** de nouveaux cas **par an**
- Cela représente **10%** de tous les diabètes mondiaux
- Le pic de survenance est chez les **enfants et adolescents avant 40 ans**

Pourquoi ces chiffres importent ? Parce que nous parlons d'enfants, de jeunes vies entièrement dépendantes d'une hormone. C'est un enjeu **majeur de santé publique**. »

## SLIDE 4 : DT1 vs DT2 - COMPARAISON FONDAMENTALE
**[Durée : 2 minutes 30 secondes]**

« Attention, point CRUCIAL : beaucoup de gens confondent Diabète Type 1 et Type 2. C'est une erreur majeure qui crée de la confusion. Laissez-moi clarifier. »

*[Pointer le tableau]*

« Regardons le tableau. La **première différence** est l'**étiologie** — la cause.

- **Type 1** : Auto-immune, destruction des cellules
- **Type 2** : Résistance insulinique, les cellules existent mais ne répondent plus

La **deuxième différence** : l'**âge**. Le Type 1 apparaît chez l'enfant ou l'adolescent. Le Type 2 chez l'adulte après 40 ans. »

« La **troisième** : l'**insuline**. Type 1 : absente. Type 2 : insuffisante mais présente. »

« Et la **prévalence** : Type 1 = 10%, Type 2 = 90%. La majorité des diabètes mondiaux est de Type 2. **Mais mon sujet, c'est le Type 1**. »

*[Souligner l'encadré rouge]*

« Pourquoi la détection précoce est-elle critique ? Parce que sans insuline, le diabétique Type 1 peut développer une **acidocétose diabétique** — un état où le corps brûle les graisses trop rapidement, créant des acides toxiques dans le sang. C'est **potentiellement fatal**, surtout chez l'enfant. »

« Et en Afrique ? Le diagnostic tardif signifie que beaucoup arrivent en urgence, déjà en acidocétose avancée. Cela augmente dramatiquement la mortalité dès la première année. »

## SLIDE 6 : SPÉCIFICITÉS AFRICAINES ET OBSTACLES DIAGNOSTIQUES
**[Durée : 3 minutes]**

« Voici LE slide le plus important pour comprendre pourquoi ma recherche est nécessaire. L'Afrique n'est pas l'Occident. La biologie, la société, les ressources — tout est différent. »

*[Pointer la colonne gauche - Formes Idiopathiques]*

« Commençons par la **biologie distinctive**. »

« En Afrique, nous voyons des **formes idiopathiques du Diabète Type 1** — c'est-à-dire des cas sans marqueurs immunologiques classiques. Qu'est-ce que cela signifie ?

- En Occident : 5 à 10% des cas sont idiopathiques
- **En Afrique : 30 à 40%** des cas sont idiopathiques
- C'est une **différence génétique fondamentale**

En plus, nous voyons une condition appelée **Ketosis-Prone Diabetes** — surnommée le « diabète africain ». C'est une forme où :

- La présentation est **aiguë** et catastrophique (pas progressive)
- Il y a une **acidocétose** (les corps s'empoisonnent)
- Paradoxalement, certains patients peuvent entrer en **rémission temporaire**
- Les mécanismes sont **complètement différents** du Type 1 classique

Maintenant, la deuxième colonne : les **obstacles diagnostiques et thérapeutiques**. C'est ici que la réalité africaine frappe dur. »

*[Énumérer lentement chaque obstacle]*

« **Obstacle 1 : Le coût des biomarqueurs.** Un test d'auto-anticorps (GADA, IA-2A, etc.) coûte entre 50 et 150 dollars américains. Le revenu mensuel moyen d'une famille camerounaise ? Environ 100 à 200 dollars. Un seul test = impossible. »

« **Obstacle 2 : Les ressources laboratoire.** La majorité des hôpitaux camerounais ne peuvent mesurer que la **glycémie capillaire** — un simple test de doigt. Pas d'auto-anticorps, pas de HbA1c sophistiquée. »

« **Obstacle 3 : Le coût de l'insuline.** Si le patient est diagnostiqué, il doit acheter de l'insuline. Celle-ci coûte 10 à 30% du revenu mensuel familial. Pour une famille pauvre, c'est **choisir entre manger et vivre**. »

« **Obstacle 4 : Les ruptures d'approvisionnement.** Même si la famille peut payer, l'insuline n'est pas toujours disponible. Les ruptures sont fréquentes et imprévisibles. »

« **Obstacle 5 : L'infrastructure.** Les laboratoires spécialisés qui font des tests avancés ? Quasi inexistants en dehors de Yaoundé et Douala. »

*[Laisser un silence significatif]*

« **Conséquence critique** : Les diagnostics arrivent souvent en **urgence**. Plus de 50% des enfants diagnostiqués en Afrique sont déjà en acidocétose avancée. En Europe ? 15 à 30%. »

« Les modèles diagnostiques occidentaux ? **Complètement inadéquats** pour cette réalité. »

« C'est ici que s'insère mon projet. Il faut une **approche africaine**, adaptée à la biologie africaine et aux ressources africaines. »

## SLIDE 13 : GESTION DÉSÉQUILIBRE DE CLASSES ET MÉTRIQUES
**[Durée : 2 minutes 30 secondes]**

« Point critique : le **déséquilibre de classes**. En médecine, c'est un piège classique. »

« Imaginez : dans ma cohorte, 90% des gens sont sains, 10% ont le diabète. Mon modèle ML peut prédire « sain » pour **tout le monde**. Qu'arrive-t-il ? »

*[Pointer colonne gauche]*

« **Accuracy** (proportion d'individus correctement classés) = 90%. Apparemment excellent ! »

« Mais **Recall** (proportion de malades détectés) = 0%. Catastrophe ! Le modèle ne détecte aucun malade. »

« C'est le piège. Une accuracy élevée masque un modèle inutile. »

*[Pointer colonne droite - Solutions]*

« **Solutions** :

1. **SMOTE** : synthétiser des instances minoritaires (créer artificiellement plus de cas malades pour équilibrer)
2. **Class weights** : donner du poids à chaque maladie détectée (pénaliser les faux négatifs)
3. **Métriques appropriées** : oublier accuracy, utiliser Recall (détection de malades), F1 (équilibre), AUC-ROC (performance globale)

Je vais utiliser ces techniques pour garantir que mon modèle détecte vraiment les à-risque. »

## SLIDE 14 : INTERPRÉTABILITÉ - SHAP vs LIME
**[Durée : 2 minutes 30 secondes]**

« Maintenant, une question éthique : comment un médecin fait-il confiance à une prédiction ML ? Il doit **comprendre pourquoi** le modèle dit « ce patient est à risque ». »

« Il y a deux approches principales : SHAP et LIME. »

*[Pointer colonne gauche - SHAP]*

« **SHAP** (Lundberg 2017) est basé sur la théorie des jeux Shapley. L'idée : chaque feature (gène, glycémie, etc.) contribue équitablement à la prédiction, comme les joueurs dans un jeu coopératif.

- Robustesse théorique solide
- État de l'art 2020-2025
- Explications globales (l'importance générale de chaque feature) ET locales (pourquoi ce patient spécifique)

*[Pointer colonne droite - LIME]*

« **LIME** (Ribeiro 2016) approxime le modèle localement avec un modèle simple et interprétable.

- Plus rapide que SHAP
- Moins robuste théoriquement
- Bon compromis vitesse/clarté
- Fonctionne avec n'importe quel modèle

*[Souligner la régulation]*

« **Régulation** : C'est important. Le RGPD (article 22) et l'EU AI Act 2026 demandent **transparence obligatoire** pour les décisions qui affectent les individus. Un diagnostic de diabète ? C'est une décision majeure. »

« Je vais utiliser **SHAP** pour garantir que mon modèle est explicable. Les médecins africains comprendront pourquoi une prédiction est faite. »

## 💡 CONSEILS POUR LA PRÉSENTATION

### Timing et Rythme
- **Total speech** : 25-28 minutes (selon rythme)
- **Pause questions** : 10 minutes
- **Grand total** : ~35-40 minutes
- **Parlez distinctement**, pas trop vite
- **Faites des pauses** après points importants

### Ton et Langage
- Restez **professionnel mais accessible**
- Utilisez des **analogies** quand c'est complexe
- Évitez le jargon excessif (définissez chaque terme)
- Montrez de l'**empathie** pour le contexte africain

### Gestures et Mouvements
- Pointez vers le tableau/slides avec **gestes amples**
- Bougez naturellement, pas de tics nerveux
- **Contact visuel** avec le jury/audience
- Position : debout, détendu, confiant
