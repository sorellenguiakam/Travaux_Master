```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Tes données
year = [2023, 2024, 2025]
Net_income = [48.3, 90.4, 249.9]

# 2. Calcul de la régression linéaire (degré 1)
# np.polyfit calcule la pente et l'ordonnée à l'origine
coefficients = np.polyfit(year, Net_income, 1)
modele = np.poly1d(coefficients)

# 3. Création de la ligne de régression
# On crée des points entre 2000 et 2002 pour tracer la droite
x_ligne = np.linspace(min(year), max(year), 10)
y_ligne = modele(x_ligne)

# 4. Affichage
plt.scatter(year, Net_income, color='blue', label='Données réelles')  # Nuage de points
plt.plot(x_ligne, y_ligne, color='red', label='Régression linéaire')  # Droite

# Ajouter des infos pour rendre le graphique propre
plt.xlabel('Année')
plt.ylabel('Revenu Net')
plt.title('Régression Linéaire du Revenu Net')
plt.legend()  # Pour afficher les étiquettes 'Données réelles' et 'Régression'

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Données basées sur l'image
Year = np.array([0, 1, 2])
Income = np.array([48.3, 90.4, 249.9])

# 2. Calcul de la régression linéaire (degré 1)
# On calcule la droite qui passe le mieux entre les points
modele = np.poly1d(np.polyfit(Year, Income, 1))

# Création d'une ligne pour le dessin (de 0 à 2)
x_trace = np.linspace(0, 2, 100)
y_trace = modele(x_trace)

# 3. Graphique
plt.figure(figsize=(8, 5))

# Tracer les points rouges ('ro' dans ton image)
plt.plot(Year, Income, 'ro', label='Données (Points)')

# Tracer la ligne de régression
plt.plot(x_trace, y_trace, 'b-', label='Régression Linéaire')

# Configuration des axes comme sur l'image
plt.xlabel('Years Since 2000')
plt.ylabel('Income (in Millions)')
plt.axis([-0.5, 2.5, 25, 275]) # [xmin, xmax, ymin, ymax]

plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()

# Afficher l'équation pour information
print(f"Équation de la droite : y = {modele.coefficients[0]:.2f}x + {modele.coefficients[1]:.2f}")
```
